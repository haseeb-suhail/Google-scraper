```python
import csv
import os
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from time import sleep
import re
from selenium.common.exceptions import NoSuchElementException

# Initialize the ChromeDriver with options and services
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument(
    '--user-agent=%s' % "Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36")
chrome_options.add_argument("--disable-blink-features=AutomationControlled")
chrome_options.add_experimental_option('useAutomationExtension', False)
chrome_options.add_experimental_option("excludeSwitches", ["enable-automation"])
# chrome_options.add_argument('profile-directory=default')

driver = webdriver.Chrome(options=chrome_options)
driver.get('https://www.google.com/advanced_search')
driver.maximize_window()
sleep(3)

exact_word = driver.find_element(By.XPATH,'//*[@id="CwYCWc"]')
exact_word.click()
exact_word.send_keys('exotic Snack')
location = driver.find_element(By.XPATH,'//*[@id="cr_button"]')
location.click()
usaa = driver.find_element(By.XPATH,'//*[@id=":7m"]')
usaa.click()
advanced_search = driver.find_element(By.XPATH,'//*[@id="s1zaZb"]/div[5]/div[8]/div[2]/input[2]')
advanced_search.click()
keyword = driver.find_elements(By.CLASS_NAME,'tjvcx')
leeen = len(keyword)
print(leeen)
for links in keyword:
    links = links.text.strip()
    print(links)
```

```python
# Collecting text from search results
keyword = driver.find_elements(By.CLASS_NAME, 'byrV5b')
leeen = len(keyword)
print(leeen)
links = []
for ele in keyword:
    link_text = ele.text
    if link_text:
        links.append(link_text)
        print(link_text)

# Save the text to a CSV file
with open('google_search_links.csv', 'a', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Links'])
    for link in links:
        writer.writerow([link])
```
